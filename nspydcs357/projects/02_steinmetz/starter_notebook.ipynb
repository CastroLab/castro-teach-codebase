# Steinmetz -- Visual Decision-Making

Neuropixels recordings during a visual discrimination task.

Reduced CSV: one row per trial with behavioral variables (contrasts, choice, feedback) and 20 PCA components of population neural activity.

```python
# Load data directly from the course repository -- no downloads needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

url = "https://raw.githubusercontent.com/CastroLab/castro-teach-codebase/main/nspydcs357/projects/02_steinmetz/data/steinmetz_reduced.csv"
df = pd.read_csv(url)
print(f"Loaded {len(df)} rows from steinmetz_reduced.csv")
```

## Explore the data

```python
# Quick look
print(f"Shape: {df.shape}")
print(f"\nBehavioral variables:")
print(df[['contrast_left','contrast_right','response','feedback']].describe().round(2))
print(f"\nAccuracy: {(df['feedback']==1).mean():.1%}")
print(f"Neural PCs: PC1 through PC{len([c for c in df.columns if c.startswith('PC')])}")
df.head()
```

## First look

```python
# Psychometric curve: accuracy as a function of contrast difference
df['contrast_diff'] = df['contrast_right'] - df['contrast_left']
grouped = df.groupby('contrast_diff')['feedback'].apply(lambda x: (x==1).mean())

plt.figure(figsize=(7, 4))
plt.bar(grouped.index, grouped.values, width=0.12, color='steelblue')
plt.xlabel('Contrast difference (right - left)')
plt.ylabel('Fraction correct')
plt.title('Psychometric curve')
plt.axhline(0.5, ls='--', color='gray', alpha=0.5)
plt.tight_layout()
plt.show()
```
