# Stringer -- 10,000 Neurons in V1

10,164 neurons recorded simultaneously during spontaneous behavior.

Reduced CSV: ~2,600 subsampled timepoints with 20 PCA components of neural activity, 3 behavioral variables (running, pupil, face motion), and 50 raw neuron traces.

```python
# Load data directly from the course repository -- no downloads needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

url = "https://raw.githubusercontent.com/CastroLab/castro-teach-codebase/main/nspydcs357/projects/06_stringer_10k/data/stringer_reduced.csv"
df = pd.read_csv(url)
print(f"Loaded {len(df)} rows from stringer_reduced.csv")
```

## Explore the data

```python
# Quick look
print(f"Shape: {df.shape}")
beh_cols = ['run_speed', 'pupil_area', 'face_motion']
pc_cols = [c for c in df.columns if c.startswith('PC')]
neuron_cols = [c for c in df.columns if c.startswith('neuron_')]
print(f"Behavioral variables: {beh_cols}")
print(f"PCA components: {len(pc_cols)}")
print(f"Raw neuron traces: {len(neuron_cols)}")
df[beh_cols].describe().round(2)
```

## First look

```python
# The key result: PC1 tracks running speed
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].scatter(df['PC1'], df['run_speed'], s=3, alpha=0.3, c='steelblue')
axes[0].set_xlabel('PC1')
axes[0].set_ylabel('Running speed')
axes[0].set_title(f"r = {df['PC1'].corr(df['run_speed']):.3f}")

axes[1].scatter(df['PC2'], df['pupil_area'], s=3, alpha=0.3, c='coral')
axes[1].set_xlabel('PC2')
axes[1].set_ylabel('Pupil area')
axes[1].set_title(f"r = {df['PC2'].corr(df['pupil_area']):.3f}")

plt.suptitle('Neural PCs track behavioral state')
plt.tight_layout()
plt.show()
```
