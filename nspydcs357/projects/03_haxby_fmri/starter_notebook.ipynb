# Haxby fMRI -- Neural Decoding

The classic 'mind-reading' fMRI dataset.

Reduced CSV: each row is one fMRI volume (non-rest), with 50 PCA components of ventral temporal cortex activity, category label, and session ID.

```python
# Load data directly from the course repository -- no downloads needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

url = "https://raw.githubusercontent.com/CastroLab/castro-teach-codebase/main/nspydcs357/projects/03_haxby_fmri/data/haxby_reduced.csv"
df = pd.read_csv(url)
print(f"Loaded {len(df)} rows from haxby_reduced.csv")
```

## Explore the data

```python
# Quick look
print(f"Shape: {df.shape}")
print(f"\nCategories:\n{df['category'].value_counts()}")
print(f"\nSessions: {df['session'].nunique()}")
print(f"PCs: 50 components (PC1-PC50)")
df.head()
```

## First look

```python
# PCA scatter: do categories separate in neural space?
from sklearn.decomposition import PCA

fig, ax = plt.subplots(figsize=(8, 6))
categories = df['category'].unique()
for i, cat in enumerate(categories):
    mask = df['category'] == cat
    ax.scatter(df.loc[mask, 'PC1'], df.loc[mask, 'PC2'],
               alpha=0.5, s=20, label=cat)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.legend(fontsize=8, ncol=2)
ax.set_title('VT cortex patterns by object category')
plt.tight_layout()
plt.show()
```
