# EEG -- Alcoholism Study

64-channel EEG from alcoholic and control subjects.

Reduced CSV: one row per trial with per-channel features (mean, std, range for 10 channels) plus subject ID and group label.

```python
# Load data directly from the course repository -- no downloads needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

url = "https://raw.githubusercontent.com/CastroLab/castro-teach-codebase/main/nspydcs357/projects/05_eeg_alcoholism/data/eeg_alcoholism_reduced.csv"
df = pd.read_csv(url)
print(f"Loaded {len(df)} rows from eeg_alcoholism_reduced.csv")
```

## Explore the data

```python
# Quick look
print(f"Shape: {df.shape}")
print(f"\nGroups:\n{df['group'].value_counts()}")
feat_cols = [c for c in df.columns if '_mean' in c or '_std' in c or '_range' in c]
print(f"\nFeature columns ({len(feat_cols)}): {feat_cols[:6]}...")
df.head()
```

## First look

```python
# Compare feature distributions by group
feat_cols = [c for c in df.columns if '_std' in c][:4]
fig, axes = plt.subplots(1, len(feat_cols), figsize=(12, 3))
for ax, feat in zip(axes, feat_cols):
    for grp, color in [('control', 'steelblue'), ('alcoholic', 'coral')]:
        ax.hist(df.loc[df['group']==grp, feat], bins=20, alpha=0.5, color=color, label=grp, density=True)
    ax.set_title(feat.replace('_std',''), fontsize=9)
    ax.legend(fontsize=7)
plt.suptitle('Voltage variability by group')
plt.tight_layout()
plt.show()
```
