# Parkinson's Telemonitoring

42 PD patients, 16 voice biomarkers, predict UPDRS severity.

Reduced CSV: one row per patient (averaged across ~140 recordings each), 13 columns after removing redundant features.

```python
# Load data directly from the course repository -- no downloads needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

url = "https://raw.githubusercontent.com/CastroLab/castro-teach-codebase/main/nspydcs357/projects/01_parkinsons/data/parkinsons_reduced.csv"
df = pd.read_csv(url)
print(f"Loaded {len(df)} rows from parkinsons_reduced.csv")
```

## Explore the data

```python
# Quick look at the data
print(f"Shape: {df.shape}")
print(f"\nTargets:")
print(df[['motor_UPDRS', 'total_UPDRS']].describe().round(2))
print(f"\nFeatures: {[c for c in df.columns if c not in ['subject#','age','sex','motor_UPDRS','total_UPDRS']]}")
df.head()
```

## First look

```python
# Correlation of voice features with disease severity
voice = [c for c in df.columns if c not in ['subject#','age','sex','motor_UPDRS','total_UPDRS']]
corr = df[voice].corrwith(df['total_UPDRS']).sort_values()

plt.figure(figsize=(8, 4))
corr.plot.barh(color='steelblue')
plt.xlabel('Correlation with total_UPDRS')
plt.title('Which voice features predict disease severity?')
plt.tight_layout()
plt.show()
```
