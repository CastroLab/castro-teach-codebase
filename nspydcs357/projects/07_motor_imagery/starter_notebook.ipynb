# Motor Imagery EEG (BCI)

64-channel EEG during motor execution and imagery.

Reduced CSV: one row per epoch with subject, run, event type, and per-channel features (alpha power, beta power, std) from 10 key channels.

```python
# Load data directly from the course repository -- no downloads needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

url = "https://raw.githubusercontent.com/CastroLab/castro-teach-codebase/main/nspydcs357/projects/07_motor_imagery/data/motor_imagery_reduced.csv"
df = pd.read_csv(url)
print(f"Loaded {len(df)} rows from motor_imagery_reduced.csv")
```

## Explore the data

```python
# Quick look
print(f"Shape: {df.shape}")
print(f"\nSubjects: {df['subject'].nunique()}")
print(f"Events:\n{df['event'].value_counts()}")
feat_cols = [c for c in df.columns if '_alpha' in c or '_beta' in c]
print(f"\nSpectral features ({len(feat_cols)}): {feat_cols[:4]}...")
df.head()
```

## First look

```python
# Alpha power by event type for motor channels
alpha_cols = [c for c in df.columns if '_alpha' in c][:4]
fig, axes = plt.subplots(1, len(alpha_cols), figsize=(12, 3))
events = sorted(df['event'].unique())
for ax, col in zip(axes, alpha_cols):
    data = [df.loc[df['event']==e, col].dropna() for e in events]
    bp = ax.boxplot(data, patch_artist=True, labels=[str(e) for e in events])
    for patch, color in zip(bp['boxes'], ['gray','steelblue','coral']):
        patch.set_facecolor(color)
    ax.set_title(col.split('_alpha')[0], fontsize=9)
    ax.set_ylabel('Alpha power')
plt.suptitle('Mu rhythm (alpha) by event type')
plt.tight_layout()
plt.show()
```
