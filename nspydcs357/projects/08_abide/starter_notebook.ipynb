# ABIDE -- Autism Brain Connectivity

Resting-state fMRI connectivity from ASD and control subjects.

Reduced CSV: one row per subject with 20 PCA components of functional connectivity, plus phenotypic data (diagnosis, age, sex, IQ).

```python
# Load data directly from the course repository -- no downloads needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

url = "https://raw.githubusercontent.com/CastroLab/castro-teach-codebase/main/nspydcs357/projects/08_abide/data/abide_reduced.csv"
df = pd.read_csv(url)
print(f"Loaded {len(df)} rows from abide_reduced.csv")
```

## Explore the data

```python
# Quick look
print(f"Shape: {df.shape}")
pheno_cols = [c for c in df.columns if not c.startswith('ConnPC')]
pc_cols = [c for c in df.columns if c.startswith('ConnPC')]
print(f"Phenotypic: {pheno_cols}")
print(f"Connectivity PCs: {len(pc_cols)}")
if 'DX_GROUP' in df.columns:
    print(f"\nDiagnosis (1=ASD, 2=Control):\n{df['DX_GROUP'].value_counts()}")
df.head()
```

## First look

```python
# Connectivity PCs by diagnosis
if 'DX_GROUP' in df.columns:
    fig, axes = plt.subplots(1, 3, figsize=(12, 3))
    for ax, pc in zip(axes, ['ConnPC1', 'ConnPC2', 'ConnPC3']):
        if pc in df.columns:
            for dx, label, color in [(1, 'ASD', 'coral'), (2, 'Control', 'steelblue')]:
                vals = df.loc[df['DX_GROUP']==dx, pc].dropna()
                ax.hist(vals, bins=10, alpha=0.5, color=color, label=label, density=True)
            ax.set_title(pc)
            ax.legend(fontsize=8)
    plt.suptitle('Connectivity PCs by diagnosis')
    plt.tight_layout()
    plt.show()
```
