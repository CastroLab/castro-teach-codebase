# Brain Tumor MRI Classification

~7,200 brain MRI scans across 4 tumor types.

Reduced CSV: one row per training image with summary statistics (mean, std, percentiles) and 50 PCA components of 64x64 grayscale pixels. Raw images available in data/raw/.

```python
# Load data directly from the course repository -- no downloads needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

url = "https://raw.githubusercontent.com/CastroLab/castro-teach-codebase/main/nspydcs357/projects/04_brain_tumor_mri/data/brain_tumor_features.csv"
df = pd.read_csv(url)
print(f"Loaded {len(df)} rows from brain_tumor_features.csv")
```

## Explore the data

```python
# Quick look
print(f"Shape: {df.shape}")
print(f"\nClasses:\n{df['label'].value_counts()}")
print(f"\nFeature columns: {[c for c in df.columns if c not in ['filename','label']][:10]}...")
df.head()
```

## First look

```python
# Do simple features distinguish tumor types?
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
for ax, feat in zip(axes, ['mean_intensity', 'std_intensity', 'p50']):
    for label in df['label'].unique():
        subset = df[df['label'] == label]
        ax.hist(subset[feat], bins=25, alpha=0.4, label=label, density=True)
    ax.set_title(feat)
    ax.legend(fontsize=7)
plt.tight_layout()
plt.show()
```
